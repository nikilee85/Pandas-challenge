```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
hero_df = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(hero_df)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#total players
totalplayers = len(purchase_data) - 1
print(totalplayers)

#sum of all prices
totalprice = purchase_data["Price"].sum()
print(totalprice)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#list of unique items
unique_items = purchase_data["Item Name"].unique()
#print(unique_items)

#number of unique itmes
unique_count = purchase_data["Item Name"].value_counts()
unique_sum = unique_count.count()
print(unique_sum)

#average price
average_price = purchase_data["Price"].mean()
print(average_price)
#summary dataframe
basic_herosummary_df = [{'Number of Unique Items': unique_sum,
                        'Average Price':"$" + str(average_price),
                        'Number of Purchases':totalplayers,
                        'Total Revenue':"$" + str(totalprice)}]

basic_df = pd.DataFrame(basic_herosummary_df)
basic_df.head()


			
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
groupedby = purchase_data.groupby(['Gender'])
gender_count = groupedby['SN'].nunique()
sum_gender_counts = gender_count.sum()


male_count = gender_count['Male']
male_percent = (male_count/sum_gender_counts) * 100


female_count = gender_count['Female']
female_percent = (female_count/sum_gender_counts) * 100

other_count = gender_count['Other / Non-Disclosed']
other_percent = (other_count/sum_gender_counts) * 100

gender_df = pd.DataFrame(gender_count)
gender_df['Percent'] = female_percent,male_percent,other_percent
gender_df["Percent"] = gender_df["Percent"].map("{:.2f}%".format)

gender_df.head()
```


## Purchasing Analysis (Gender)

```python
#average_price_per = 
price_df = pd.DataFrame({'Purchase Count': purchase_count,
             'Average Price': + sorted_average_price,
             'Total Price': sorted_price,
             'Average Price Per Person': average_price_per})

price_df.head()

```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#creating bins and cut
bins = [0,9,13,18,23,28,33,38,100]	

group_names = ["<10", "10-14", "15-19", "20-24","25-29","30-34", "35-39","40+"]

purchase_data["Age Groups"] = pd.cut(purchase_data["Age"],bins,labels=group_names,include_lowest=True)

age_group_value = purchase_data["Age Groups"].value_counts()
age_group_value.head()
value_percent = (age_group_value / unique_sum) * 100
value_percent.head()
value_percent.map("{:.2f}%".format)

#print(age_group_value)
#print(value_percent)
age_group_df = pd.DataFrame({'Total Count': age_group_value,
            'Percent of Players': + value_percent,
           })

age_group_df.sort_index()

```
